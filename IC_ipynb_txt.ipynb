#Regressão e regularização

Estabeleceremos alguns modelos de regressão linear, com e sem regularização ou padronização, e compará-los.

##Primeira parte: com padronização

Nessa parte vamos olhar para os modelos quando utilizamos a padronização.

Vamos separar nossos dados em subconjuntos de teste e treino, respectivos tanto a esta quanto à segunda parte. 

```
features = dataframe.drop(columns = ['PRICE'])
target = dataframe['PRICE']
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size =0.2, random_state = 42)

# criando cópias para poder criar um modelo sem padronização
X_train9 = X_train.copy()
X_test9 = X_test.copy()
y_train9 = y_train.copy()
y_test9 = y_test.copy()

# iniciando metodo pra padronizar
scalar = StandardScaler()
```

Alguns hiperparâmetros serão testados para termos uma ideia de qual será um valor ótimo de predição do nosso modelo, dado um determinado valor de $\alpha$. Vamos testar um total de 12 hiperparâmetros.

```
# Candidatos para alpha(lambda)
hyper_alpha = [0.15,0.20,0.25,0.5,0.55,0.6,0.7,0.75,1,1.25,1.5,2]
```

Primeiramente, vamos testar as regressões com regularização e dados padronizados em *x_train2* e *y_train2*, iterando sobre todos os hiperparâmetros e armazenando os seus respectivos MSE's e $R^2$ para analisá-los.

```
# iniciando listas para lasso e ridge que irão guardar erro quadratico, hyperparametro e o R^2
teste_l = []
treino_l = []
teste_r = []
treino_r = []

# inciando loop
for i in range(len(hyper_alpha)-1, -1, -1):

  X_train2, X_val, y_train2, y_val = train_test_split(X_train, y_train, test_size = 0.2, random_state = 42)
  # adequando e transformando os dados de treino com parametros estimados em X_train
  no_binary_train = X_train2.drop(columns = ['CHAS'])
  no_binary_test = X_val.drop(columns = ['CHAS'])

  X_train2 = scalar.fit_transform(no_binary_train)
  X_val = scalar.transform(no_binary_test)
  
  X_train2 = no_binary_train.assign(CHAS = dataframe['CHAS'])
  X_val = no_binary_test.assign(CHAS = dataframe['CHAS'])

  # Criando instancias de regresao linear Lasso, Ridge
  lasso1 = Lasso(alpha = hyper_alpha[i])
  ridge1 = Ridge(alpha = hyper_alpha[i])

  # Regularização Lasso
  lasso1.fit(X_train2, y_train2)
  treino_predito_l = lasso1.predict(X_train2)
  teste_predito_l = lasso1.predict(X_val)

  # Regularzação Ridge
  ridge1.fit(X_train2, y_train2)
  treino_predito_r = ridge1.predict(X_train2)
  teste_predito_r = ridge1.predict(X_val)

  # adicionar as predicoes nas suas respectivas listas
   
  treino_l.append([mse(y_train2, treino_predito_l), r(y_train2, treino_predito_l)])
  teste_l.append([mse(y_val, teste_predito_l), r(y_val, teste_predito_l)])

  treino_r.append([mse(y_train2, treino_predito_r), r(y_train2, treino_predito_r)])
  teste_r.append([mse(y_val, teste_predito_r), r(y_val, teste_predito_r)])
```

Agora, faremos a mesma coisa porém com um modelo sem regularização e utilizando outro conjunto de treino, também iterando sobre os valores de $\alpha$. 

```
# iniciando listas para lasso e ridge que irão guardar erro quadratico, hyperparametro e o R^2
teste_lr2 = []
treino_lr2 = []
teste_l2 = []
treino_l2 = []
teste_r2 = []
treino_r2 = []
```

```
# inciando loop
for i in range(len(hyper_alpha)-1,-1,-1):

  #X_train, X_tes, y_train, y_test = train_test_split(X_train, y_train, test_size = 0.2, random_state = 42)
  # adequando e transformando os dados de treino com parametros estimados em X_train
  no_binary_train2 = X_train.drop(columns = ['CHAS'])
  no_binary_test2 = X_test.drop(columns = ['CHAS'])

  X_train = scalar.fit_transform(no_binary_train2)
  X_test = scalar.transform(no_binary_test2)
  
  X_train = no_binary_train2.insert(3, "CHAS", dataframe['CHAS'])
  X_test = no_binary_test2.insert(3, "CHAS", dataframe['CHAS'])

  print(X_train)
  #perfomando a regressão linear
  linear_model = LinearRegression()
  linear_model.fit(X_train, y_train)
  modelo_linear_predito = linear_model.predict(X_train)
  modelo_linear_preditot = linear_model.predict(X_test)

  # Criando instancias de regresao linear Lasso, Ridge
  lasso2 = Lasso(alpha = hyper_alpha[i])
  ridge2 = Ridge(alpha = hyper_alpha[i])

  # Regularização Lasso
  lasso2.fit(X_train, y_train)
  treino_predito_l2 = lasso2.predict(X_train)
  teste_predito_l2 = lasso2.predict(X_test)

  # Regularzação Ridge
  ridge2.fit(X_train, y_train)
  treino_predito_r2 = ridge2.predict(X_train)
  teste_predito_r2 = ridge2.predict(X_test)

  # adicionar as predicoes nas suas respectivas listas
  treino_lr2.append([mse(y_train, modelo_linear_predito), r(y_train, modelo_linear_predito)])
  teste_lr2.append([mse(y_test, modelo_linear_preditot), r(y_test, modelo_linear_preditot)])

  treino_l2.append([mse(y_train, treino_predito_l2),r(y_train, treino_predito_l2)])
  teste_l2.append([mse(y_test, teste_predito_l2), r(y_test, teste_predito_l2)])

  treino_r2.append([mse(y_train, treino_predito_r2), r(y_train, treino_predito_r2)])
  teste_r2.append([mse(y_test, teste_predito_r2),  r(y_test, teste_predito_r2)])  
```

Podemos observar que o hiperparâmetro **não** altera os valores de MSE e $R^2$ da regressão sem regularização, já que esse método não depende de nenhum parâmetro externo. 

É notável também que, quanto menor nosso hiperparâmetro, menor também será nosso MSE e naturalmente maior será nosso $R^2$. Isso provavelmente se dá pois, além de o nosso *dataset* ser bem preparado, por nosso espaço de hipóteses ser limitado à funções lineares, a regularização faz com que o modelo não se aplique tão bem aos dados como faria a *best fitting line*. ##Segunda parte: sem a padronização das variáveis

Depois de analisarmos os modelos para os casos com a padronização, vamos analisar como as regressões se comportam sem a padronização dos dados.

Faremos exatamente como no segundo caso da primeira parte, porém, comentando as linhas de *fit* e *transform*, e adicionando um "9" nas variáveis para diferenciá-las.

```
# Repetir as partes de regressão sem a padronização das variáveis e comparar os resultados
# iniciando listas para lasso e ridge que irão guardar erro quadratico, hyperparametro e o R^2
teste_lr9 = []
treino_lr9 = []
teste_l9 = []
treino_l9 = []
teste_r9 = []
treino_r9 = []
# inciando loop
for i in range(len(hyper_alpha)-1,-1,-1):

  #X_train, X_tes, y_train, y_test = train_test_split(X_train, y_train, test_size = 0.2, random_state = 42)
  # adequando e transformando os dados de treino com parametros estimados em X_train
  #X_train = scalar.fit_transform(X_train)
  #X_test = scalar.transform(X_test)
  
  #perfomando a regressão linear
  linear_model9 = LinearRegression()
  linear_model9.fit(X_train9, y_train9)
  modelo_linear_predito9 = linear_model9.predict(X_train9)
  modelo_linear_preditot9 = linear_model9.predict(X_test9)

  # Criando instancias de regresao linear Lasso, Ridge
  lasso9 = Lasso(alpha = hyper_alpha[i])
  ridge9 = Ridge(alpha = hyper_alpha[i])

  # Regularização Lasso
  lasso9.fit(X_train9, y_train9)
  treino_predito_l9 = lasso9.predict(X_train9)
  teste_predito_l9 = lasso9.predict(X_test9)

  # Regularzação Ridge
  ridge9.fit(X_train9, y_train9)
  treino_predito_r9 = ridge9.predict(X_train9)
  teste_predito_r9 = ridge9.predict(X_test9)

  # adicionar as predicoes nas suas respectivas listas
  treino_lr9.append([mse(y_train9, modelo_linear_predito9), r(y_train9, modelo_linear_predito9)])
  teste_lr9.append([mse(y_test9, modelo_linear_preditot9), r(y_test9, modelo_linear_preditot9)])

  treino_l9.append([mse(y_train9, treino_predito_l9),r(y_train9, treino_predito_l9)])
  teste_l9.append([mse(y_test9, teste_predito_l9), r(y_test9, teste_predito_l9)])

  treino_r9.append([mse(y_train9, treino_predito_r9), r(y_train9, treino_predito_r9)])
  teste_r9.append([mse(y_test9, teste_predito_r9),  r(y_test9, teste_predito_r9)])  
```
